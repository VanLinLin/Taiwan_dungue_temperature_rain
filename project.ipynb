```python
import pandas as pd
import numpy as np
```

```python
# 讀取登革熱資料
dungue = pd.read_csv(r"D:\Site_project\Dengue\dungue.csv")
```

```python
# 刪除登革熱資料多於欄位
dungue = dungue[['Date_Confirmation', 'County_living', 'Number_of_confirmed_cases']]
# 修改格式
dungue = dungue.astype({'Date_Confirmation': 'string', 'County_living': 'string'})
# 找出None位置並刪除該列
dungue.drop(axis=0, index=dungue[dungue['Date_Confirmation'].values == 'None'].index, inplace=True)
dungue.drop(axis=0, index=dungue[dungue['County_living'].values == 'None'].index, inplace=True)
# 重置索引
dungue.reset_index(drop=True, inplace=True)
# 將確診數量缺失值設為0
dungue.fillna({'Number_of_confirmed_cases':0})
dungue
```
